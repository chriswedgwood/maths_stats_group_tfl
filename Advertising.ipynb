```python
import pandas as pd
from sklearn import linear_model
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_advertising = pd.read_csv('Advertising.csv', usecols=["TV","radio","newspaper","sales"])
```

```python
df_advertising[["TV","radio","newspaper"]]
```

```python
x = df_advertising[["TV","radio","newspaper"]]
y = df_advertising[["sales"]]
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
```

```python
predictions = lm.predict(x)
print(predictions[0:5])
```

```python
lm.score(x,y)
```

Above is the R² score of our model. As you probably remember, this the percentage of explained variance of the predictions

```python
lm.coef_
```

```python
lm.intercept_
```
